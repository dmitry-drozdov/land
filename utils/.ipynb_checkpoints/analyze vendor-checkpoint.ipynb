```python
import os
```

```python
main_root = 'e:\\phd\\test_repos_calls\\'
```

```python
vendors = set()
vendors_by_repo = {}
```

```python
for root, dirs, files in os.walk(main_root):
    if 'vendor' not in dirs:
        continue
    if 'results' in root or 'LICENSES' in root:
        continue
    rf = root + '\\vendor'
    if not os.path.isdir(rf):
        continue
    print(rf)
    for sub_folder in os.listdir(rf):
        rfs = rf + '\\' + sub_folder
        if not os.path.isdir(rfs):
            continue
        for subsub_folder in os.listdir(rfs):
            rfss = rfs + '\\' + subsub_folder
            if os.path.isdir(rfss):
                path = rfss.split('\\')
                vendor = path[-2]+'\\'+path[-1]
                print('\t', vendor)
                vendors.add(vendor)
                if path[-2] in vendors_by_repo:
                    vendors_by_repo[path[-2]]+=1
                else:
                    vendors_by_repo[path[-2]]=1
```

```python
print(len(vendors))
```
